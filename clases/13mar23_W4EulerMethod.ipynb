```python
import matplotlib.pyplot as plt
```

```python
import math
#ODE
def dydt(t,s):
  return math.exp(-t)
```

```python
h = 0.1
t = 0
tf = 1+h 
s0 = -1
s = s0
```

```python
sol = [s] # we need a list to keep track of the path
ta = [t]
while(t<tf):
  s = s + h*dydt(t, s)
  t = t + h
  sol.append(s)
  ta.append(t)
```

```python
sol_exact = [-math.exp(-t) for t in ta]
```

```python
plt.figure(figsize = (12, 8))
plt.plot(ta, sol, 'bo--', label='Approximate')
plt.plot(ta, sol_exact, 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
