```python
plt.figure(figsize = (12, 8))
plt.plot(h_list,err, 'bo--', label='Error vs h');
plt.title('Error en el metodo numerico como funcion del paso de integracion para dxdt=2t')
plt.xlabel('h')
plt.ylabel('Error')
plt.grid()
plt.legend(loc='lower right')
```
